```python
t_env = t.filterSamples(lambda x, y, z: True if z["study"] != "NSIP" else False)
```

```python
def get_raref_vector(data, stride=10, iters=25):
    rare_data = vstack([[len(h.nonzero()[0]) for h in rarefaction(data, stride=stride)] \
                        for i in range(iters)])
    X = arange(stride,data.sum()+stride,stride)
    return X, mean(rare_data, axis=0)
```

```python
dfs = {}
for sid in t_env.SampleIds:
    counts = t.sampleData(sid)
    x, raref_data = get_raref_vector(counts)
    dfs[sid] = pd.DataFrame(data={"x" : x, "y" : raref_data})   
```

```python
d = pd.concat(dfs, axis=0)
d.reset_index(inplace=True)
d.drop(["level_1"], axis=1, inplace=True)
d.rename(columns={"level_0" : "SampleID"}, inplace=True)
d.head()
```

```python
def get_hist(counts):
    h_full = histogram(counts, bins=arange(counts.min(), counts.max()+2))
    z = h_full[0]!=0
    h_no_zeros = [i[z] for i in h_full]
    return [i[1:] for i in h_no_zeros]

def print_hist(fn, h):
    with open(fn, "w") as out:
        for c, i in zip(h[0], h[1]):
            out.write("%s,%s\n"%(int(i),c))
            
def catchall_wrap(fn, outfile):
    !CatchAllCmdL.exe $fn $outfile >/dev/null 2>&1
    
def parse_best_model_stats(fn):
    line_count = 0
    for line in open(fn):
        if line_count == 0:
            total_sp = int(line.split(",")[0].split("=")[1])
            line_count += 1
            continue
        try:
            if line.startswith("Best Parm Model"):
                stats = line.rstrip().split(",")
                est_sp = float(stats[4])
                lcb = float(stats[6])
                ucb = float(stats[7])
                break
        except IndexError:
            continue
        if line.startswith("Parm Model 2a"):
            stats = line.rstrip().split(",")
            est_sp = float(stats[4])
            lcb = float(stats[6])
            ucb = float(stats[7])
            break
    return total_sp, est_sp, lcb, ucb

def full_catchall_analysis(sid, data):
    outdir = !mktemp -d
    outdir = outdir[0]
    h = get_hist(data)
    print_hist(outdir + "/input.csv", h)
    catchall_wrap(outdir + "/input.csv", outdir)
    total_sp, est_sp, lcb, ucb = parse_best_model_stats(outdir + "/input_BestModelsAnalysis.csv")
    return {sid : {"total_sp" : total_sp,
                   "est_sp" : est_sp,
                   "lcb" : lcb,
                   "ucb": ucb}}
```

```python
catchall_dict = {}
for s, data in zip(t.SampleIds, t.iterSampleData()):
    catchall_dict.update(full_catchall_analysis(s, data))
```
